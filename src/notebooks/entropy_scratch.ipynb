```python
from scipy.stats import entropy

ents = []
for bin in sorted(df_zf.density_bin.value_counts().keys()):
    # append the result
    cpy_arr = cpy_bins[cpy_bins["density_bin"] == bin]["citations_per_year"].values
    ents.append(entropy(cpy_arr))

ents
```

```python
import plotnine as pn

bins = sorted(df_zf.density_bin.value_counts().keys())
bins = [float(item.split(",")[0].replace("(", "")) for item in bins]
df = pd.DataFrame(
    data=zip(bins, df_zf.density_bin.value_counts().sort_index(), ents),
    columns=["density_bin", "count", "citation entropy"],
)

(
    pn.ggplot(
        df, 
    )
    + pn.geom_point(pn.aes(
        x="density_bin",
        fill="citation entropy",
        y="count",
    ),size=10,
    )
    + pn.labs(fill="citation entropy,\nH(CPY | density_bin)\n")
)
```
